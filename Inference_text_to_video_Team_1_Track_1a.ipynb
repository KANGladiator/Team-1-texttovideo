```python
!pip install gradio
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video
from PIL import Image

pipe = DiffusionPipeline.from_pretrained("cerspense/zeroscope_v2_576w", torch_dtype=torch.float16)
pipe.enable_model_cpu_offload()

# Memory optimization
pipe.unet.enable_forward_chunking(chunk_size=1, dim=1)
pipe.enable_vae_slicing()
```

```python
import gradio as gr
import os

# Directory to save generated videos
video_folder = os.path.join('content', 'Videos')
os.makedirs(video_folder, exist_ok=True)

def generate_video(prompt):
    # Generate the video frames
    video_frames = pipe(prompt, num_frames=24).frames[0]

    # Save the video in the folder
    video_name = f"{prompt.replace(' ', '_')}.mp4"
    video_path = os.path.join(video_folder, video_name)
    export_to_video(video_frames, video_path)

    # Return the same path for both displaying and downloading the video
    return video_path, video_path

# Define Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## 🎥 Sora like Video Generator pipeline")

    # Team details
    gr.Markdown("""
    ### Team 1:
    - **Aditya Bharadwaj**: SE21UARI007
    - **Ashwin Baduni**: SE21UCAM003
    - **Kanishk Singh**: SE21UCAM005
    - **Yashaswi Singh**: SE21UCAM017
    """)

    with gr.Row():
        prompt_input = gr.Textbox(label="Enter your prompt", placeholder="Describe your scene...")

    with gr.Row():
        generate_btn = gr.Button("Generate")
        download_btn = gr.Button("Download")

    video_output = gr.Video(label="Generated Video", interactive=False)

    # Link button action to function
    generate_btn.click(
        fn=generate_video,
        inputs=[prompt_input],
        outputs=[video_output, download_btn],  # Two outputs: Video and File
    )

# Launch Gradio interface
demo.launch(debug=True)
```
